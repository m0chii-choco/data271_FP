## Function how it works

1. rec_inc: What this function does is, it has two arguments: code1, code2. Where code1 and code2 are the letter codes of the continent it makes two new data sets of code1 and code2. Then it makes a scatter plot where the x axis is the plastic waste recycled and the y axis is the plastic waste that is incinerated, and it mapped to the color of the continent. 
2. lml: for does the exact same as rec_inc; the only difference is that it takes the columns of Share of littered and mismanaged from total regional waste and Share of waste landfilled from total regional waste.

````python
" Code_pick: string -> plot"
"code pick expects a string of the entiy of the code and returns a plot that shows the rec and inc waste"

def rec_inc(code1, code2):
    codex1 = spf.loc[spf['Code'].str.contains(code1)]
    codex2 = spf.loc[spf['Code'].str.contains(code2)]
    fig = plt.figure()
    plt.title('Share Plastic waste recyced & incinerated from total regional waste')
    sns.scatterplot(codex1, x='Share of waste recycled from total regional waste', y ='Share of waste incinerated from total regional waste',
                    label = code1)
    sns.scatterplot(codex2, x='Share of waste recycled from total regional waste', y ='Share of waste incinerated from total regional waste',
                    label = code2)
rec_inc('USA', 'OCE')
````

````python
def lml (code, code2 ,df):
    codex = df.loc[df['Code'].str.contains(code)]
    codex2 = df.loc[df['Code'].str.contains(code2)]
    fig = plt.figure()
    plt.title('Share Plastic waste Mismanaged & landfill from total regional waste')
    sns.scatterplot(codex, x='Share of littered and mismanaged from total regional waste', y ='Share of waste landfilled from total regional waste',
                          label = code)
    sns.scatterplot(codex2, x='Share of littered and mismanaged from total regional waste', y ='Share of waste landfilled from total regional waste',
                          label = code2)
lml('USA', 'OCE',spf)
````

````python
fig, ax = plt.subplots(2, 2, figsize = (10,6))
sns.histplot(spf, x='Share of waste recycled from total regional waste', ax = ax[0,0])
sns.histplot(spf, x ='Share of waste incinerated from total regional waste', ax = ax[0,1])
sns.histplot(spf, x ='Share of littered and mismanaged from total regional waste',ax = ax[1,0])
sns.histplot(spf, x ='Share of waste landfilled from total regional waste',ax = ax[1,1])
plt.show()
````
